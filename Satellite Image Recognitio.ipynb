```python
#importing libraries
```

```python
import numpy as np
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
#uploading training set and performing data augmentation
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)
training_set = train_datagen.flow_from_directory(
    'training_set',target_size=(64,64),batch_size=32,class_mode='categorical')
```

```python
#uploading validation set and performing data augmentation
```

```python
validation_datagen = ImageDataGenerator(rescale=1./255)
validation_set = validation_datagen.flow_from_directory(
    'validation_set',target_size=(64,64),batch_size=32,class_mode='categorical')
```

```python
#uploading test set and performing data augmentation
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
    'test_set',target_size=(64,64),batch_size=32,class_mode='categorical')
```

```python
#importing sequential model
```

```python
cnn = tf.keras.models.Sequential()
```

```python
#adding first convolutional layer
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=64,kernel_size=2,activation='relu',input_shape=[64,64,3]))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
#adding second convolutional layer
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=64,kernel_size=2,activation='relu',input_shape=[64,64,3]))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
#adding dropout rate of 0.5
```

```python
cnn.add(tf.keras.layers.Dropout(0.5))
```

```python
#flattening the input before sending it to dense layer
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
#adding dense layers
```

```python
cnn.add(tf.keras.layers.Dense(units=124,activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=4,activation='softmax'))
```

```python
#compiling model
```

```python
cnn.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#training the model with validation set
```

```python
history=(cnn.fit(x=training_set, validation_data=validation_set,epochs=10))
```

```python
#testing the model with test set
```

```python
history=(cnn.fit(x=training_set, validation_data=test_set,epochs=5))
```

```python
#plotting graph of accuracy throughout epochs
```

```python
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.6, 1])
plt.legend(loc='lower left')
```

```python
#plotting graph of loss throughout epochs
```

```python
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('model loss')
plt.ylim([0, 1])
plt.legend(loc='lower left')
```

```python
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['validation'], loc = 'upper left')
plt.show()
```
